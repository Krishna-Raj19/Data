```python
import pandas as pd
import os
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
fifa=pd.read_csv('C:/Users/Krishna Raj/Desktop/Numpy/fifa.csv')
fifa
```

```python
rc=['ID','Name','Age','Overall','Value','Wage','International Reputation','Position','Joined','Contract Valid Until','Height','Weight','Release Clause']
```

```python
df=fifa.loc[:,rc]
df
```

```python
a=df.Wage.str.replace('€','')
```

```python
def cc(value):    
    v = value[-1]
    if v == 'K':
        value = value[:-1]
        value = float(value) * 1000
    elif v == 'M':
        value = value[:-1]
        value = float(value) * 1000000
    
    else:
        value = float(value)
        
    return value
```

```python
df.Wage=a.apply(cc)
df
```

```python
b=df.Value.str.replace('€','')
```

```python
df.Value=b.apply(cc)
df['Value']
```

```python
c=df['Release Clause'].str.replace('€','')
```

```python
d=c.fillna(method='ffill').str[1:]
d
```

```python
df['Contract Valid Until'] = pd.to_datetime(df['Contract Valid Until'],errors='coerce').fillna(method='ffill')
df['Contract Valid Until']
```

```python
a = df['Joined'].str[-4:].astype(float)
a
```

```python
joined_mean = int(round(a.mean()))
joined_mean
```

```python
df['Joined']=df.Joined.str[-4:].fillna(joined_mean).astype(int)
df['Joined']
```

```python
df['Weight'] = df['Weight'].fillna(method = 'ffill').str[:-3].astype(float)
df['Weight']
```

```python
df['Height'] = df['Height'].str.replace("'",'.').astype(float)
df['Height']
```
